```python
import matplotlib.pyplot as plt, seaborn as sns, warnings, random, pandas as pd
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=[16,5])
x = [random.randrange(100) for i in range(100)]
y = [random.randrange(100) for i in range(100)]
plt.hist([x, y])
plt.show()
```

```python
plt.figure(figsize=[16,5])
plt.hist([x, y], color=['r','b'], alpha=0.5)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 5))
for a in [x, y]:
    sns.distplot(a, bins=range(1, 110, 10), ax=ax, kde=False, color='g')
ax.set_xlim([0, 100])
plt.show()
```
